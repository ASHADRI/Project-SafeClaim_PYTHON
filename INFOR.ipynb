- Months_As_Customer número de meses que el cliente lleva asegurado.
- Age edad del cliente.
- Policy_Number identificador único de la póliza.
- Policy_Type tipo de póliza (Personal Auto, Corporate Auto, etc.).
- Policy_Annual_Premium valor anual de la prima del seguro.
- Vehicle_Category tipo de vehículo (Sedan, Sports Car, SUV, etc.).
- Vehicle_Price rango de precio del vehículo.
- Vehicle_Claim monto del reclamo presentado.
- Auto_Make marca del vehículo (Toyota, Honda, Ford, etc.).
- Auto_Model modelo del vehículo.
- Driver_Rating calificación de conducción del asegurado (1–5).
- Past_Number_of_Claims número de reclamos anteriores del cliente.
- Number_of_Suppliments número de suplementos o adiciones a la póliza.
- Address_Change_Claim si el cliente cambió de dirección recientemente (YesNo).
- Days_Policy_Accident días entre el inicio de la póliza y el accidente.
- Days_Policy_Claim días entre el inicio de la póliza y la reclamación.
- Age_of_Vehicle antigüedad del vehículo (en años).
- Age_of_Policy_Holder antigüedad del titular de la póliza (en años).
- Police_Report_Available si existe reporte policial (YesNo).
- Witnesses número de testigos declarados.
- Fault quién tuvo la culpa (Policy Holder  Third Party).
- Accident_Area zona donde ocurrió el siniestro (Urban  Rural).
- Fraud_Found_P variable objetivo → 1 si el reclamo es fraudulento, 0 si es legítimo.
